# Visualize dataset

Explore dataset to see how it looks like and more importantly, check if the data loaders work as expected!

```python
import os
import numpy as np
from visualization import display_images, display_image_masks
from data.loaders import train_validation_loaders, test_loader
```

```python
DATASET_DIR = './dataset'
BATCH_SIZE=4
train_dir = os.path.join(DATASET_DIR,'stage1_train')
test_dir = os.path.join(DATASET_DIR,'stage1_test')
```

Create data loaders:

```python
train_loader, validation_loader = train_validation_loaders(train_dir,validation_ratio=0.1,batch_size=BATCH_SIZE,num_workers=1)
```

```python
print("Dataset object:")
print(train_loader.dataset)
print("# batches: %s" % len(train_loader))
print("# validation batches: %s" % len(validation_loader))
```

```python
test_loader = test_loader(test_dir,batch_size=BATCH_SIZE)
```

```python
print("Test dataset:")
print(test_loader.dataset)
print("# batches: %s" % len(test_loader))
```

Now, lets check a few images in the train dataset

```python
train_img, train_masks, train_boxes, train_classes, train_names = next(iter(train_loader))
```

```python
display_images(train_img)
```

And look at the masks:

```python
display_image_masks(train_img[0,:,:,:],train_masks[0,:,:,:],train_boxes[0],title=train_names[0])
```

What about test set?

```python
test_img, test_names = next(iter(test_loader))
```

```python
display_images(test_img)
```

Note that test images are resized. After running them through the model for inference, we have to undo the changes to get actual coordinates of the masks.
